```
import xml.etree.ElementTree as ET
import pandas as pd 
import numpy as np

data = 'parkinglot.xml'
parking = ET.parse(data).getroot()

columns = [
        "SpaceID", 
        "Occupied",
        "SizeX", 
        "SizeY", 
        "SizeW", 
        "SizeH",
        "Angle", 
        "PointAX", 
        "PointAY", 
        "PointBX", 
        "PointBY", 
        "PointCX", 
        "PointCY", 
        "PointDX", 
        "PointDY" 
        ]
    
df = pd.DataFrame(columns=columns)

for space in parking.iter('space'):
    entry = []

    id = space.attrib.get('id')
    occupied = space.attrib.get("occupied")

    entry += [id, occupied]

    for rotrec in space.iter("rotatedRect"):
        for center in rotrec.iter("center"):
            x = center.attrib.get("x")
            y = center.attrib.get("y")
        for size in rotrec.iter("size"):
            w = size.attrib.get("w")
            h = size.attrib.get("h")
        for angle in rotrec.iter("angle"):
            d = angle.attrib.get("d")

        entry += [x, y, w, h, d]
    
    for contour in space.iter("contour"):

        for point in contour.iter("point"):
            x = point.attrib.get("x")
            y = point.attrib.get("y")

            entry += [x, y]

    df.loc[len(df)] = entry

df.to_csv("parkinglot.csv", encoding='utf-8', index=False)
```
